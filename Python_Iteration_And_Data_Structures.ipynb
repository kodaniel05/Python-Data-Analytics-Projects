## Question 2

```python
# Part a
for year in range(1800,1851):
    yearDF = pd.read_csv(f"/anvil/projects/tdm/data/noaa/{year}.csv", header = None, names=["id","date","element_code","value","mflag","qflag","sflag","obstime"])
    avg = yearDF[yearDF["element_code"] == "PRCP"]["value"].mean()
    print(f"Year: {year} Average Precipitation: {avg}")
    
```

```python
# Part b
year = 1800
while year < 1851:
    yearDF = pd.read_csv(f"/anvil/projects/tdm/data/noaa/{year}.csv", header = None, names=["id","date","element_code","value","mflag","qflag","sflag","obstime"])
    avg = yearDF[yearDF["element_code"] == "PRCP"]["value"].mean()
    print(f"Year: {year} Average Precipitation: {avg}")
    if avg > 22:
        break;
    year += 1
    
```
